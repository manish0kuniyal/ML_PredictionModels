```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('road.jfif')  # Replace with your image path

# Convert the image to the HSV color space
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Define the lower and upper bounds for blue color in HSV
lower_blue = np.array([100, 50, 50])
upper_blue = np.array([130, 255, 255])

# Create a mask using the inRange function
blue_mask = cv2.inRange(hsv_image, lower_blue, upper_blue)

# Create an all-black image with the same shape as the original image
result_image = np.zeros_like(image)

# Set the blue regions in the result image to be the original color, rest to black and white
result_image[blue_mask > 0] = image[blue_mask > 0]

# Display the result
plt.imshow(cv2.cvtColor(result_image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('road2.jfif')  

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(gray_image, 100, 200)
contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
edge_marked_image = image.copy()

cv2.drawContours(edge_marked_image, contours, -1, (10, 0, 205), 2)

plt.subplot(1, 3, 3) 
plt.imshow(cv2.cvtColor(edge_marked_image, cv2.COLOR_BGR2RGB))
plt.show()
```
