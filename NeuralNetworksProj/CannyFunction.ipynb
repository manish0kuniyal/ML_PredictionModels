```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('road3.jpg')

lower_green = np.array([0, 100, 0])
upper_green = np.array([100, 255, 100])

lower_red = np.array([0, 0, 100])
upper_red = np.array([100, 100, 255])

hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

green_mask = cv2.inRange(hsv, lower_green, upper_green)
red_mask = cv2.inRange(hsv, lower_red, upper_red)

highlighted_image = cv2.merge((green_mask, red_mask, red_mask))

plt.imshow(cv2.cvtColor(highlighted_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('road3.jpg')
road_color = [112, 160, 175]

lower_bound = np.array(road_color, dtype=np.uint8) - np.array([30, 30, 30])
upper_bound = np.array(road_color, dtype=np.uint8) + np.array([30, 30, 30])
road_mask = cv2.inRange(image, lower_bound, upper_bound)

other_mask = cv2.bitwise_not(road_mask)

highlighted_image = cv2.merge([other_mask, road_mask, other_mask], image)

plt.imshow(cv2.cvtColor(highlighted_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
